### Utilize algum teste de hipótese para identificar qual a possível distribuição das colunas da questão anterior

```python
import scipy.special as scsp
def zScoreToValue(z):
  return 0.5 * (1 + scsp.erf(z/np.sqrt(2)))
```

```python
hipotese_longitude = longi.groupby(['longitude']).size().reset_index(name='Fabs')
hipotese_longitude['Fac'] = hipotese_longitude['Fabs'].cumsum() 
hipotese_longitude['Frac'] = hipotese_longitude['Fac'].max()
hipotese_longitude['Zi'] = hipotese_longitude['longitude'].apply(lambda x: (x-media_longitude)/std_longitude)
hipotese_longitude['FracEsp'] = hipotese_longitude['Zi'].apply(lambda x: zScoreToValue(x))
hipotese_longitude['D_negativo'] = abs(hipotese_longitude['FracEsp'])-hipotese_longitude['Frac']
hipotese_longitude['D_positivo'] = 0
for i in range(hipotese_longitude['Frac'].shape[0]):
  if i > 0:
    hipotese_longitude['D_positivo'].iloc[i] = hipotese_longitude['FracEsp'].iloc[i] - hipotese_longitude['Frac'].iloc[i-1]
  else:
    hipotese_longitude['D_positivo'].iloc[i] = hipotese_longitude['FracEsp'].iloc[i]
hipotese_longitude
```

```python
D = (hipotese_longitude[['D_positivo','D_negativo']].max()).max()
from scipy.stats import ksone

def ks_critical_value(n_trials, alpha):
  return ksone.ppf(1-alpha/2, n_trials)

p_value = ks_critical_value(longi.shape[0],0.05)

if D < p_value:
  print("Os dados de longitude seguem uma distribuição normal")
else:
  print("Os dados de longitude não seguem uma distribuição normal")
```
